```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(link)
```

```python
df.head()
```

```python
df=df.dropna()
```

```python
train_input=np.array(df['x'][0:500]).reshape(500,1)
train_output=np.array(df['y'][0:500]).reshape(500,1)
test_input=np.array(df['x'][500:699]).reshape(199,1)
test_output=np.array(df['y'][500:699]).reshape(199,1)
```

```python
def forward_propagation(train_input,parameters):
    m=parameters['m']
    c=parameters['c']
    predictions=(train_input*m)+c
    return predictions
```

```python
def cost_function(predictions,train_output):
    cost=np.mean((predictions-train_output)**2)/2
    return cost
```

```python
def backward_propagation(predictions,train_input,train_output):
    derivatives=dict()
    df=predictions-train_output
    dm=np.mean(np.multiply(df,train_input))
    dc=np.mean(df)
    derivatives['dm']=dm
    derivatives['dc']=dc
    return derivatives
```

```python
def update_parameters(parameters,derivatives,learning_rate):
    parameters['m']-=learning_rate*derivatives['dm']
    parameters['c']-=learning_rate*derivatives['dc']
    return parameters
```

```python
def train(train_input,train_output,learning_rate,it):
    parameters=dict()
    parameters['m']=np.random.uniform(0,1)
    parameters['c']=np.random.uniform(0,1)
    
    loss=list()
    
    for i in range(it):
        predictions=forward_propagation(train_input,parameters)
        cost=cost_function(predictions,train_output)
        loss.append(cost)
        print(f'it : {i+1} , cost : {cost}')
        
        plt.plot(train_input,train_output,'+',label='Original')
        plt.plot(train_input,predictions,'*',label='Training')
        plt.legend()
        plt.show()
        
        derivatives=backward_propagation(predictions,train_input,train_output)
        parameters=update_parameters(parameters,derivatives,learning_rate)
        
    return parameters, loss
```

```python
parameters,loss=train(train_input,train_output,0.0001,20)
```
